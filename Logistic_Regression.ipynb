```python
# import packages
import numpy as np
import matplotlib.pyplot as plt 

import pandas as pd  
import seaborn as sns 
%matplotlib inline
```

```python
# Excel数据导入
data = pd.read_excel("数据3.xlsx")
print ("Number of Records: ", data.shape[0])
data.head()
```

```python
# 选定分隔符将列表或元组转换为字符串，将一个特征变量变为计算机能读懂的特征距离
data = data.join(pd.get_dummies(data['famtype'], prefix = 'famtype'))  
data = data.drop(['famtype'], axis = 1)  

data = data.join(pd.get_dummies(data['expectation'], prefix = 'expectation'))  
data = data.drop(['expectation'], axis = 1)  

data = data.join(pd.get_dummies(data['socistatu'], prefix = 'socistatu'))  
data = data.drop(['socistatu'], axis = 1)    

data = data.drop(['name', 'tel', 'recptime', 'recpplace', 'remark'], axis=1)

data.head()
```

```python
# 因变量：分布情况
print(data['entrepre'].unique())  # 返回参数组中所有不同的值，并按从小到大排列.里面的参数为待预测字段
data_label_count = data.groupby('entrepre').count()['gender'].sort_values(ascending = False)  # 将DataFrame分组后，分别对cc栏计数，并排序
print(data_label_count)
data_label_count.plot.bar()
plt.show()
```

```python
# 自变量：分布情况&相关性分析
graph_by_variables = data.columns
plt.figure(figsize = (15, 18))
for i in range(0, 28):
    plt.subplot(7, 4, i+1)
    sns.distplot(data[graph_by_variables[i]])
    plt.title(graph_by_variables[i])
plt.tight_layout()  # 自动调整子图参数，使之填满图像区域
```

```python
f, ax = plt.subplots(figsize = (15, 15))
sns.heatmap(data.corr(), annot = True, linewidths = 0.5, fmt = '.1f', ax = ax)
```

```python
# 数据集划分
from sklearn.model_selection import train_test_split
# Splitting into train and test sets
X = data.drop(['entrepre'], axis = 1)
y = data['entrepre']
# test_size = 0.2意为训练集占80%，测试集占20%，即将数据集按4:1的比例划分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 114514)
print(len(X_train), len(X_test))
```

```python
from sklearn.linear_model import LogisticRegression
# Fitting a logistic regression model with default parameters
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# 评价
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
```

```python
# Prediction & Evaluation
print("Logistic regression score for test set:")
# Precision, Recall, F1-score
y_hat_test = logreg.predict(X_test)
print("Precision: {:.3f}".format(precision_score(y_test, y_hat_test)), "Recall: {:.3f}".format(recall_score(y_test, y_hat_test)))
# 精确率是针对预测结果而言的，它表示的是预测为正的样本中有多少是真正的正样本（测试集预测的准确性）。
# 召回率是针对原来的样本而言的，它表示的是样本中的正例有多少被预测正确了（样本集预测的准确性）。
print("F1 score: {:.3f}".format(f1_score(y_test, y_hat_test)))
# 综合评价指标是精确率和召回率的加权调和平均，更具有综合代表性。
# AUC
y_hat_test = logreg.predict_proba(X_test)
print("AUC score: {:.3f}".format(roc_auc_score(y_test, y_hat_test[:,1])))
# AUC越大，分类器分类效果越好
```

```python
def preProcessing(data):
    # 选定分隔符将列表或元组转换为字符串，将一个特征变量变为计算机能读懂的特征距离
    data = data.join(pd.get_dummies(data['famtype'], prefix='famtype'))
    data = data.drop(['famtype'], axis=1)

    data = data.join(pd.get_dummies(data['expectation'], prefix='expectation'))
    data = data.drop(['expectation'], axis=1)

    data = data.join(pd.get_dummies(data['socistatu'], prefix='socistatu'))
    data = data.drop(['socistatu'], axis=1)

    data = data.drop(['name', 'tel', 'recptime', 'recpplace', 'remark'], axis=1)

    return data

data = pd.read_excel('D:\SUFE\Projects\VS\FarmerStartUpIndicator\数据3.xlsx')
data = preProcessing(data)
data.head()
```

```python
def Logistic_Regression(data):
    # 数据集划分
    X = data.drop(['entrepre'], axis=1)
    y = data['entrepre']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=114514)
    # print(len(X_train), len(X_test))

    # 训练模型
    logreg = LogisticRegression()
    logreg.fit(X_train, y_train)

    return logreg
```
